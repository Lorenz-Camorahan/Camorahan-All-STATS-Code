```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# load dataset
df = pd.read_csv(r'/Users/lorenzcamorahan/Desktop/Statistics/current-power.csv', 
                delimiter=",")

df
```

```python
df.describe()
```

```python
x = df['Current']
y = df['Power']

plt.scatter(x, y)
plt.title('Device Current Vs Power Response')
plt.xlabel('Current (A)')
plt.ylabel('Power (W)')
plt.grid()
```

```python
df.cov()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color='blue',
            alpha=0.6,
            label = "Data Points")
plt.plot(x, line,
            color='red',
            label=f'Best Fit Line(r = {r_value:.3f})')
plt.grid()
plt.xlabel('Current(A)')
plt.ylabel('Power(W)')
plt.title('Linear Regression: Device Current Vs Power Response')
plt.legend()
plt.show
```

## Air Quality <hr style = "border:2.5px solid #0000FF"></hr>

```python
air = pd.read_csv(r'/Users/lorenzcamorahan/Desktop/Statistics/air-quality.csv', delimiter=",")
air
```

## NOx Vs Relative_Humidity Correlation

```python
#regression line
x = air['NOx']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color='black',
            alpha=0.2,
            label = "Data Points")
plt.plot(x, line,
            color='red',
            label=f'Best Fit Line(r = {r_value:.3f})')
plt.grid()
plt.xlabel('NOx (ug/cu.m)')
plt.ylabel('Relative Humidity (%)')
plt.title('NOx Vs Humidity Correlation')
plt.legend()
plt.show
```

## NO2 Vs Humidity Correlation

```python
#regression line
x = air['NO2']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color='black',
            alpha=0.2,
            label = "Data Points")
plt.plot(x, line,
            color='red',
            label=f'Best Fit Line(r = {r_value:.3f})')
plt.grid()
plt.xlabel('NO2 (ug/cu.m)')
plt.ylabel('Relative Humidity (%)')
plt.title('NO2 Vs Humidity Correlation')
plt.legend()
plt.show
```
